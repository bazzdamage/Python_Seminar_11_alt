```python
import sympy
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x', real=True)
```

```python
y = x**2 - 6*abs(x) + 8
y
```

```python
solve(y, x)
```

```python
import matplotlib.pyplot as pl

f = plot(y)

print(f)
```

```python
y1 = diff(y)
y1
print(f'судя по всему - вершина = {y.subs(x,0)}')
```

```python
# минимум x
x_min = solve(y1, x)[0]
# минимум y
y_min = y.subs(x, solve(y1, x)[0])
print(f'минимум функции = x: {x_min}, y: {y_min}')
```
